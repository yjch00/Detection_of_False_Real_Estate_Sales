```python
def to01(aaa):
    if aaa == '전세':
        return 0
    else:
        return 1
def to02(aaa):
    if aaa == 'aaaaaaaaa':
        return 1
    else:
        return 0
def to03(aaa):
    if aaa == 'aaaaaaaaa':
        return 0
    else:
        return 1
```

```python
b_m2['위치'] = b_m2['시군구']+' '+b_m2['번지']
b_m2['전월세'] = list(map(to01,b_m2['전월세구분']))
b_m2['신축정도'] = b_m2['건축년도']
b_m2['면적']=b_m2.iloc[:,3]
b_m3 = b_m2.drop([b_m2.columns[0],b_m2.columns[1],b_m2.columns[2],b_m2.columns[3]],axis=1)
b_m3['보증금'] = b_m3['보증금(만원)']
b_m3['월세'] = b_m3['월세(만원)']
```

```python
b_o2['위치'] = b_o2['시군구']+' '+b_o2['번지']
b_o2['전월세'] = list(map(to01,b_o2['전월세구분']))
b_o2['신축정도'] = b_o2['건축년도']
b_o2['면적']=b_o2.iloc[:,3]
b_o3 = b_o2.drop([b_o2.columns[0],b_o2.columns[1],b_o2.columns[2],b_o2.columns[3]],axis=1)
b_o3['보증금'] = b_o3['보증금(만원)']
b_o3['월세'] = b_o3['월세(만원)']
```

```python
b_m3 = b_m3[['unique', '위치', '면적', '전월세', '보증금','월세','층', '건축년도', '계약년월']]
b_o3 = b_o3[['unique', '위치', '면적', '전월세', '보증금','월세','층', '건축년도', '계약년월']]
```

```python
b_m3['type']=list(map(to02,b_m3['전월세']))
b_o3['type']=list(map(to03,b_o3['전월세']))
```

```python
b_m3 = b_m3[['unique', '위치', '면적', '전월세', '보증금','월세','층', '건축년도', '계약년월', 'type']]
b_o3 = b_o3[['unique', '위치', '면적', '전월세', '보증금','월세','층', '건축년도', '계약년월', 'type']]
```

```python
data = pd.concat([b_m3, b_o3])
```

```python
def no_c(aaa):
    num = ''
    for one in aaa:
        if one != ',':
            num += one
    return float(num)
```

```python
def no_c(aaa):
    num = ''
    for one in str(aaa):
        if one != ',':
            num += one
    return float(num)
data['월세'] = list(map(no_c,data['월세']))
```

```python
def per(a,b,c):
    return(a+5*float(b))/c
data['perarea'] = list(map(per,data['보증금'], data['월세'],data['면적']))
```
